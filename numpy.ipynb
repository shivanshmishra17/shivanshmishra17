```python
# Sine and consine of values
a = np.array([90,0,60,30])
b = np.radians(a)  # b = np.deg2rad(a)
print(b)

c = np.sin(b)
d = np.cos(a, dtype = np.float16)
print(c)
print(d)
```

```python
#Matrix Multiplication
a = np.ones((2,3))
print(a)
b = np.full((3,2),2)
print(b)
c = np.matmul(a,b)
print(c)

c= c.astype('int16')
print(c)
```

```python
#Transpose of a matrix
c = np.array([[4,5,2],[2,9,7],[6,5,8]])
print(c)

d = np.transpose(c)
print(d)
```

```python
# Determinant of a matrix
a = np.identity(3)
print(a)
b =np.linalg.det(a)
print(b)
```

```python
a = np.array([[1,2,3],[4,5,6]])
print(a)

#Minimum Element
minimum = np.min(a)
print(minimum)
minimum = np.min(a, axis=1)
print(minimum)

#Maximum Element
maximum = np.max(a)
print(maximum)

#Sum of elements
add = np.sum(a,axis = 1)
print(add)

#Mean of elements
avg = np.mean(a)
print(avg)

#Median of matrix
med = np.median(a)
print(med)
```

5. Write a NumPy program to test a given array element-wise for finiteness (not infinity or not a Number).

```python
arr1 = np.array([1,2,3,4,5])
print(np.isfinite(arr1))
arr2 = np.array([1,2,np.nan,np.inf,2,3,4])
print(np.isfinite(arr2))
```

8. Write a NumPy program to test element-wise for complex number, real number of a given array. Also test whether a given number is a scalar type or not.

```python
arr = np.array([1,3+0j,6+3j,2,8,7+4j,9])
print('Checking for complex elements:',np.iscomplex(arr))
print('Checking for real elements:',np.isreal(arr))
print('Checking if the element is scalar or not:',np.isscalar(2))
print('Checking if the element is scalar or not:',np.isscalar(arr[3]))
```
